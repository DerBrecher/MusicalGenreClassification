# Helper notebook used for image generation and testing

```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np

filepath = "Full_On_clips/02 - Organic Soup - People Of The Earth (Scopelyser Remix)- Clip1.wav"

data, sr = librosa.load(filepath, sr=44100)
S = librosa.feature.mfcc(y=data, sr=sr, n_fft=2048, hop_length=1024)

fig, ax = plt.subplots(figsize=[12,7])
#S = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S, x_axis='time', sr=sr,
                         fmax=8000, ax=ax);
plt.ylabel("MFCC")
```
